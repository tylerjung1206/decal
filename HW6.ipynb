```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
list1 = np.random.randint(0,101,40)
list2 = np.random.randint(0,101,40)
list3 = np.random.randint(0,101,40)

x1 = np.arange(40)
fig, axs = plt.subplots(2)
axs[0].plot(x1,list1, color = "orange", linewidth = 10, label = 'list1')
axs[0].plot(x1,list2, color = "red", linestyle = "dashdot", label = 'list2')
axs[0].set_xlabel("position in list")
axs[0].set_ylabel("random int")
axs[0].set_title("list 1 and 2")
axs[0].legend()
axs[1].scatter(x1,list3, marker = "D", color = "magenta")
axs[1].set_xlabel("position in list")
axs[1].set_ylabel("random int")
axs[1].set_title("list3")
plt.show()
```

```python
#2
P,V = np.meshgrid(np.linspace(1,10,100),np.linspace(10,30))
T = (P + 16.02 / (V ** 2)) * (V - 0.1124) / 0.083144
plt.figure()
plt.title("van der Waals")
plt.imshow(T, cmap = 'plasma', extent = [1,10,10,30], vmin = 300, vmax = 3000)
cbar = plt.colorbar()
cbar.set_label("Temperature")
plt.xlabel("Pressure (bar)")
plt.ylabel("Volume (liters)")
plt.show
```

```python
Ns = [1e3, 1e4, 1e5, 1e6]

def funcpi(N):
    points = np.random.rand(int(N), 2)
    distances = np.sqrt(points[:, 0]**2 + points[:, 1]**2)
    points_inside = distances[distances <= 1]
    N_inside = len(points_inside)
    
    pi_est = 4 * (N_inside / N)
    
    return pi_est, points

for i in Ns:
    pi_est, _ = funcpi(N)
    print(f" {i} : estimation = {pi_est}")

N2 = 1e4
_, points = funcpi(N2)
plt.figure()
plt.scatter(points[:, 0], points[:, 1], c =(points[:, 0]**2 + points[:, 1]**2 <= 1), cmap = "magma", vmin = -1.5, vmax = 1.5)
plt.title("Est of Pi with 1e4")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
